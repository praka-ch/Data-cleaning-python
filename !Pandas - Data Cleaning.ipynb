```python
import pandas as pd
```

```python
df = pd.read_excel(r"D:\! DA\Pandas Tutorial\Customer Call List.xlsx")
df
```

# Duplicate Data

```python
#create duplicate data (!IMPORTANT ALWAYS DUPLICATES DATA)

df = df.drop_duplicates()
df
```

# Deleting Column

```python
#remove/discard unnecessary columns (Not_Useful_Column)

df = df.drop(columns = "Not_Useful_Column")
df
```

# Filtering Data

```python
#remove/discard unnecessary string (Last_Name) 

df["Last_Name"] = df["Last_Name"].str.strip("123./_")
df
```

# Replace Data

```python
#delete unnecessary any characters and equalize all data (Phone_Number)

df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '', regex = True)
df
```

# Change String

```python
#convert the data type to a string 

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

# Add New Character

```python
#add a character to the data column

df['Phone_Number'] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
#get rid of unnecessary data (nan-- , Na--)

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

# Split Column

```python
#separates a column into new columns

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand = True)
df
```

# Identification String as Same All

```python
#identify strings to be all the same

df["Paying Customer"] = df["Paying Customer"].replace({'Y' : 'Yes', 'N' : 'No'}, regex = False)
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({'Y' : 'Yes', 'N' : 'No'}, regex = False)
df
```

# Drop Null Data

```python
#removes all data that is null

df = df.fillna("")
df
```

# Job Task

Kasus = Data yang ditampilkan hanya pengguna yang tidak ingin ditelpon 

```python
#eliminates data in the Do_Not_Contact column that has the value Yes

for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Yes':
        df.drop(x, inplace = True)

df
```

```python
#eliminates data in the Phone_Number column that has the value Blank

for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace = True)

df

#Another way to drop null values !RECOMMENDED 
#df = df.dropna(subset = "Phone_Number", inplace = True)
```
